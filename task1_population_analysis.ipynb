```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load dataset
file_path = r"C:\Users\Lenovo\OneDrive\Desktop\task1-prodigy\API_SP.POP.TOTL_DS2_en_csv_v2_38144.csv.csv"
df = pd.read_csv(file_path, skiprows=4)

# Choose year (change if needed)
year = '2020'

# Keep relevant columns and drop missing values
population_data = df[['Country Name', 'Country Code', year]].dropna()

# Keep only real countries (Country Code = 3 letters)
population_data = population_data[population_data['Country Code'].str.len() == 3]

# Sort and get top 10 countries
top10 = population_data.sort_values(by=year, ascending=False).head(10)

# ========================
# 📊 HORIZONTAL BAR CHART
# ========================
plt.figure(figsize=(10,6))
plt.barh(top10['Country Name'], top10[year])
plt.title(f"Top 10 Most Populated Countries ({year})")
plt.xlabel("Population")
plt.ylabel("Country")

plt.gca().invert_yaxis()  # Highest population on top
plt.tight_layout()
plt.show()

# ========================
# 📊 HISTOGRAM (LOG SCALE)
# ========================
plt.figure(figsize=(8,5))
plt.hist(np.log10(population_data[year]), bins=20)

plt.title(f"Log Distribution of Population Across Countries ({year})")
plt.xlabel("Log10(Population)")
plt.ylabel("Number of Countries")
plt.grid()

plt.tight_layout()
plt.show()
```
